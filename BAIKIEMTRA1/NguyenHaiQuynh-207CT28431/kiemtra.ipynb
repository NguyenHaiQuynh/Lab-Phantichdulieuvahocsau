```python
# 1 Tạo một DataFrame từ dữ liệu trên 
import pandas as pd

names = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ivan", "Jack", "Kelly", "Liam", "Mona", "Nina", "Oscar"]
ages = [25, 30, 35, 28, 22, 45, 34, 31, 27, 29, 33, 40, 26, 32, 36]
salary = [50000, 60000, 70000, 55000, 52000, 80000, 72000, 68000, 61000, 59000, 63000, 77000, 53000, 66000, 75000]

data = {
    'Name': names,
    'Age': ages,
    'Salary': salary
}

df = pd.DataFrame(data)
# 2 Hiển thị thông tin về DataFrame vừa tạo
print(df)
```

```python
#3 Lọc các hàng trong DataFrame có 'Age' lớn hơn 28
filtered_df = df[df['Age'] > 28]
print(filtered_df)
```

```python
#4 Tính giá trị trung bình của cột 'Salary'
df2 = df['Salary'].mean()
print("Giá trị trung bình của cột Salary là:", df2)
```

```python
#5 Nhóm dữ liệu theo cột 'Age' và tính tổng 'Salary' cho mỗi nhóm
group_df = df.groupby('Age')['Salary'].sum().reset_index()
print(group_df)
```

```python
# 6 Sắp xếp DataFrame theo cột 'Salary' giảm dần
df_Sap_Xep_Data = df.sort_values(by='Salary', ascending=False)
```

```python
print(df_Sap_Xep_Data)
```

```python
#7 Vẽ biểu đồ cột cho cột 'Age'
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Age'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age of Individuals')
plt.xticks(rotation=45) 
plt.show()
```

```python
#8 Vẽ biểu đồ đường cho cột 'Salary'
plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Salary'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary of Individuals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 9 Vẽ biểu đồ tròn cho cột 'Age'
plt.figure(figsize=(8, 8))
plt.pie(df['Age'], labels=df['Name'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Age Distribution Among Individuals')
plt.axis('equal')  # Đảm bảo biểu đồ đường tròn được vẽ tròn
plt.show()
```

```python
# 10 Vẽ biểu đồ phân tán cho 'Age' và 'Salary'
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], color='b', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot of Age vs Salary')
##Thêm nhãn cho từng điểm dữ liệu
for i in range(len(df)):
    plt.text(df['Age'][i], df['Salary'][i], df['Name'][i], fontsize=9, ha='right')

plt.grid(True)
plt.show()
```

```python
# 11 Kiểm tra xem có giá trị NaN nào trong DataFrame không
has_nan = df.isna().any().any()
print("DataFrame có giá trị NaN:", has_nan)
```

```python
# 12 Thay thế các giá trị của cột 'Age' lớn hơn 30 bằng giá trị trung bình của cột đó
mean_age = df['Age'].mean()
#Thay thế các giá trị của cột 'Age' lớn hơn 30 bằng giá trị trung bình
df.loc[df['Age'] > 30, 'Age'] = mean_age
print(df)
```

```python
# 13 Chuẩn hóa(normalize) cột 'Age' về khoảng giá trị từ 0 đến 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
print(df)
```

```python
# 14 Tạo một cột mới 'Age_group' phân loại tuổi thành 'Young','Middle-aged' và 'Old' dựa trên giá trị của cột 'Age'
bins = [0, 30, 45, 100]
labels = ['Young', 'Middle-aged', 'Old']
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

print(df)
```

```python
#15 Tính toán tỷ lệ phần trăm thay đổi (percentage change) của cột 'Salary'
df['Salary_pct_change'] = df['Salary'].pct_change() * 100
print(df)
```

```python
#16 Tìm các giá trị trùng lặp trong DataFrame dựa trên cột 'Name' và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên
df.drop_duplicates(subset=['Name'], keep='first', inplace=True)
print(df)
```
